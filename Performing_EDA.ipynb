````python
df['purchase_Year-Month'] = df['Purchase Date'].dt.to_period('M') # converts date time col in df to period
````

````python
purchase_trend = df.groupby('purchase_Year-Month').size()
````

```plt.figure(figsize=(12, 6))``` : Creates a new figure with the size of 12 inches in width and 6 inches in height. This adjusts the overall dimensions of the graph for better visibility.


- ```purchase_trend.plot```: Plots the data from the purchase_trend dataframe.
- ```kind='line'```: Specifies that it is a line plot.
- ```marker='o'```: Adds circular markers at each data point.
- ```color='blue'```: Sets the line and marker color to blue.







````python
plt.figure(figsize=(12, 6))
purchase_trend.plot(kind='line', marker='o', color='blue')
plt.title('Number of Purchases Over Time')
#Sets the title of the plot to "Number of Purchases Over Time."
plt.xlabel('Year-Month')
plt.ylabel('Number of Purchases')
plt.grid(True)
#Enables gridlines on the plot to make it easier to read and align data points with the axis.
plt.xticks(rotation=45)
#Rotates the x-axis labels by 45 degrees for better readability, especially when the labels are longer (e.g., months or dates).
plt.show()
````
